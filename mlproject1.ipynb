```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

```python
y=df['logS']
y
```

```python
x=df.drop('logS',axis=1,)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test,=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

##**Model building**

### **Linear regression**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

###**model performance**

```python
from sklearn.metrics import mean_squared_error,r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)

lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
print('LR MSE (Train):',lr_train_mse)
print('LR R2 (Train):',lr_train_r2)
print('LR MSE (Test):',lr_test_mse)
print('LR R2 (Test):',lr_test_r2)
```

```python
lr_results=pd.DataFrame(['Linear regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
```

```python
lr_results
```

##**forest model**

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

```python
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

###**model performance**

```python
from sklearn.metrics import mean_squared_error,r2_score
rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2=r2_score(y_train,y_rf_train_pred)

rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
```

```python
rf_results=pd.DataFrame(['Random forest',rf_train_mse,rf_train_r2,lr_test_mse,rf_test_r2]).transpose()
rf_results.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
rf_results
```

##**MODEL COMPARISON**

```python
df_models=pd.concat([lr_results,rf_results],axis=0)
df_models
```

```python
df_models.reset_index(drop=True)
```

##**Data visualisation**

```python
import matplotlib.pyplot as plt
import seaborn as sns
print('Visualization libraries imported successfully.')
```

```python
plt.figure(figsize=(10, 5))

# Linear Regression Plot
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
sns.scatterplot(x=y_test, y=y_lr_test_pred, alpha=0.6)
plt.plot([min(y_test.min(), y_lr_test_pred.min()), max(y_test.max(), y_lr_test_pred.max())], [min(y_test.min(), y_lr_test_pred.min()), max(y_test.max(), y_lr_test_pred.max())], color='red', linestyle='--')
plt.xlabel('Experimental logS')
plt.ylabel('Predicted logS')
plt.title('Linear Regression: Experimental vs. Predicted logS')

# Random Forest Plot
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
sns.scatterplot(x=y_test, y=y_rf_test_pred, alpha=0.6)
plt.plot([min(y_test.min(), y_rf_test_pred.min()), max(y_test.max(), y_rf_test_pred.max())], [min(y_test.min(), y_rf_test_pred.min()), max(y_test.max(), y_rf_test_pred.max())], color='red', linestyle='--')
plt.xlabel('Experimental logS')
plt.ylabel('Predicted logS')
plt.title('Random Forest: Experimental vs. Predicted logS')

plt.tight_layout() # Adjust layout to prevent overlapping titles and labels
plt.show()
```

```python
df_melted = df_models.reset_index(drop=True).melt(id_vars=['Method'], var_name='Metric', value_name='Score')
df_melted.head()
```

```python
plt.figure(figsize=(12, 6))

# MSE comparison plot
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
sns.barplot(x='Metric', y='Score', hue='Method', data=df_melted[df_melted['Metric'].isin(['Training MSE', 'Test MSE'])])
plt.title('Model MSE Comparison')
plt.xlabel('Metric')
plt.ylabel('Score')
plt.legend(title='Model')

# R2 comparison plot
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
sns.barplot(x='Metric', y='Score', hue='Method', data=df_melted[df_melted['Metric'].isin(['Training R2', 'Test R2'])])
plt.title('Model R2 Comparison')
plt.xlabel('Metric')
plt.ylabel('Score')
plt.legend(title='Model')

plt.tight_layout()
plt.show()
```
